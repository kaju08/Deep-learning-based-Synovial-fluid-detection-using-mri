```python
import os
import shutil
import pandas as pd
import random
from PIL import Image
from glob import glob
from sklearn.model_selection import train_test_split
seed = 42
random.seed(seed)
```

```python
base_dir = "synovium dataset with mask images"
```

```python
read_path_list = []

for label in os.listdir(base_dir):
    label_folder = os.path.join(base_dir, label)
    sub_folders = os.listdir(label_folder)
    
    for sub_folder in sub_folders:
        pat_folders_path = os.path.join(label_folder, sub_folder)
        pat_folders = os.listdir(pat_folders_path)
        
        for pat_folder in pat_folders:
            present_pat_path = os.path.join(pat_folders_path, pat_folder)
            present_pat_subfolders = os.listdir(present_pat_path)
            
            for pat_subfolder in present_pat_subfolders:
                if str(pat_subfolder).lower().startswith("m"):
                    pat_subfolder_path = os.path.join(present_pat_path, pat_subfolder)
                    all_files = os.listdir(pat_subfolder_path)
                    all_files.sort()
                    for file_ in all_files:
                        read_path = os.path.join(pat_subfolder_path, file_)
                        read_path_list.append(read_path)
                elif str(pat_subfolder).lower().startswith("s") or str(pat_subfolder).lower().startswith("2") or str(pat_subfolder).lower().startswith("1"):
                    pat_subfolder_path = os.path.join(present_pat_path, pat_subfolder)
                    all_files = os.listdir(pat_subfolder_path)
                    all_files.sort()
                    for file_ in all_files:
                        read_path = os.path.join(pat_subfolder_path, file_)
                        read_path_list.append(read_path)
                else:
                    print(f"{os.path.join(present_pat_path, pat_subfolder)}")
                    print('Expected directory not found')
```

```python
len(read_path_list)
```

```python
df = pd.DataFrame()
df['path'] = read_path_list

df.head()
```

```python
pat_name = []
image_status = []
target = []
sub_folder = []
for idx, values in df.iterrows():
    pat_name.append(values['path'].split("\\")[-3].lower())
    image_status.append(values['path'].split("\\")[-2].lower())
    target.append(values['path'].split("\\")[1].lower())
    sub_folder.append(values['path'].split("\\")[2].lower())
```

```python
df['pat_name'] = pat_name
df['image_status'] = image_status
df['target'] = target
df['sub_folder'] = sub_folder

df.head()
```

```python
df['image_status'].unique()
```

```python
df['image_status'] = df['image_status'].apply(lambda x: 'mask' if x.lower().startswith('m') else 'saggital')
```

```python
df['sub_folder'].unique()
```

```python
df['pat_name'] = df['pat_name'].apply(lambda x: ''.join(x.split(" ")))
```

```python
df['pat_name'].unique()
```

```python
sy1_status_ = []
sy2_status_ = []

mask_sy1 = []
image_sy1 = []

mask_sy2 = []
image_sy2 = []

for target_ in df['target'].unique():
    print(f"{target_}")
    target_df = df.loc[df['target'] == target_]

    for pat_name_ in df['pat_name'].unique():
        target1 = target_df.loc[target_df['pat_name'] == pat_name_]

        synovial1 = target1.loc[target1['sub_folder'] == f'{target_}-synovial']
        mask_sy1_count = len(synovial1.loc[synovial1['image_status'] == 'mask'])
        image_sy1_count = len(synovial1.loc[synovial1['image_status'] == 'saggital'])
        
        sy1_status_.append(os.path.join(target_, f'{target_}-synovial', pat_name_))
        mask_sy1.append(mask_sy1_count)
        image_sy1.append(image_sy1_count)
        
        synovial2 = target1.loc[target1['sub_folder'] == f'{target_}-synovial2']
        mask_sy2_count = len(synovial2.loc[synovial2['image_status'] == 'mask'])
        image_sy2_count = len(synovial2.loc[synovial2['image_status'] == 'saggital'])
        
        sy2_status_.append(os.path.join(target_, f'{target_}-synovial2', pat_name_))
        mask_sy2.append(mask_sy2_count)
        image_sy2.append(image_sy2_count)
```

```python
new_df = pd.DataFrame()

new_df['status'] = sy1_status_
new_df['mask_sy1'] = mask_sy1
new_df['image_sy1'] = image_sy1
new_df = new_df[(new_df['mask_sy1'] != 0) | (new_df['image_sy1'] != 0)]
```

```python
base_dir = "dataset"
output_dir = "segricated"

if output_dir in os.listdir("."):
    shutil.rmtree(output_dir)
    os.makedirs(os.path.join(output_dir, 'mask'))
    os.makedirs(os.path.join(output_dir, 'original'))
else:
    os.makedirs(os.path.join(output_dir, 'mask'))
    os.makedirs(os.path.join(output_dir, 'original'))
```

```python
for label in os.listdir(base_dir):
    label_folder = os.path.join(base_dir, label)
    sub_folders = os.listdir(label_folder)
    
    for sub_folder in sub_folders:
        pat_folders_path = os.path.join(label_folder, sub_folder)
        pat_folders = os.listdir(pat_folders_path)
        
        for pat_folder in pat_folders:
            present_pat_path = os.path.join(pat_folders_path, pat_folder)
            present_pat_subfolders = os.listdir(present_pat_path)
            
            for pat_subfolder in present_pat_subfolders:
                if str(pat_subfolder).lower().startswith("m"):
                    pat_subfolder_path = os.path.join(present_pat_path, pat_subfolder)
                    all_files = os.listdir(pat_subfolder_path)
                    all_files.sort()
                    for idx, file_ in enumerate(all_files):
                        mask_image = Image.open(os.path.join(pat_subfolder_path, file_))
                        mask_image = mask_image.resize((224, 224), resample=Image.Resampling.BICUBIC)
                        mask_image.save(os.path.join(output_dir, 'mask', f'{label}_{pat_folder}_{idx}.png'))

                elif str(pat_subfolder).lower().startswith("s") or str(pat_subfolder).lower().startswith("2") or str(pat_subfolder).lower().startswith("1"):
                    pat_subfolder_path = os.path.join(present_pat_path, pat_subfolder)
                    all_files = os.listdir(pat_subfolder_path)
                    all_files.sort()
                    for idx, file_ in enumerate(all_files):
                        original_image = Image.open(os.path.join(pat_subfolder_path, file_))
                        original_image = original_image.resize((224, 224), resample=Image.Resampling.BICUBIC)
                        original_image.save(os.path.join(output_dir, 'original', f'{label}_{pat_folder}_{idx}.png'))

                else:
                    print(f"{os.path.join(present_pat_path, pat_subfolder)}")
                    print('Expected directory not found')
```
